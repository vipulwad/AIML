## Group by operations involves some combination of spliting applying a function applying a function and combining the results
- This can be used to group large amounts of data and compute operations on this group

```python
merge_df.groupby(by="sex")['charges in INR'].sum()
#based on 'sex' grouping, it displays sum for column 'charges in INR'
```

```python
merge_df.groupby(by="region")['age'].mean()
```

```python
merge_df.groupby(by="sex")['bmi'].median()
```

## Aggregate Function

```python
# find min and max age, median bmi, mean of children , count of smoker for all male and female records
merge_df.groupby("sex").agg({'age': ['min','max'], 'bmi':'median', 'smoker':'count', 'children':'mean'})
```

```python
merge_df['children'].mean()
```

```python
merge_df['smoker']=merge_df['smoker'].map({'yes':1, 'no':0})
```
