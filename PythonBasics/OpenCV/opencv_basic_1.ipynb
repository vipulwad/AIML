# OpenCV
Compared to other languages like C/C++, Python is slower. But another important feature of Python is that it can be easily extended with C/C++. This feature helps us to write computationally intensive codes in C/C++ and create a Python wrapper for it so that we can use these wrappers as Python modules.

#### This gives us two advantages:Â 
* First, our code is as fast as original C/C++ code (since it is the actual C++ code working in background) andÂ 
* Second, it is very easy to code in Python. This is how OpenCV-Python works, it is a Python wrapper around original C++ implementation.

And the support of Numpy makes the task easier. All the OpenCV array structures are converted to-and-from Numpy arrays.Â

## Convert RGB to HSV, HLS
* An RGB (colored) image has three channels, Red, Blue and Green. A colored image in OpenCV has a shape in [H, W, C] format, where H, W, and C are image height, width and number of channels. All three channels have a value range between 0 and 255.

* The HSV image also has three channels, the Hue, Saturation and Value channels. In OpenCV, the values of the Hue channel range from 0 to 179, whereas the Saturation and Value channels range from 0 to 255.

* An HLS image also has three channels, the Hue, Lightness and Saturation channels. In OpenCV, the values of the Hue channel range from 0 to 179 whereas the Lightness and Saturation channels range from 0 to 255.

## Resize image - resize()
```
cv2.resize(source, dsize, dest, fx, fy, interpolation)
```
* source: Input Image array (Single-channel, 8-bit or floating-point) 
* dsize: Size of the output array
* dest: Output array (Similar to the dimensions and type of Input image array) [optional]
* fx: Scale factor along the horizontal axis  [optional]
* fy: Scale factor along the vertical axis  [optional]
* interpolation: One of the above interpolation methods  [optional]
  * cv2.INTER_AREA: This is used when we need to shrink an image.
  * cv2.INTER_CUBIC: This is slow but more efficient.
  * cv2.INTER_LINEAR: This is primarily used when zooming is required. This is the default interpolation technique in OpenCV.

````python
img = cv2.imread('/content/drive/MyDrive/IMARTICUS/IPYNB/opencv/openCV-data/chicky_512.png', 1)
````

````python
resize_img = cv2.resize(img, (600,400)) # Provide in WxH
print(resize_img.shape) # Returns HXW
plt.imshow(resize_img)
````

````python
resize_shrink = cv2.resize(img, (100,100)) # Provide in WxH
print(resize_shrink.shape) # Returns HXW
plt.imshow(resize_shrink)
````

### Resize image by ratio

````python
ratio_resize = cv2.resize(img, (0,0),  fx=0.13, fy=0.75)

plt.imshow(ratio_resize)
print('Image size: ',ratio_resize.shape)
````

````python
img.shape
````

### Resize image by interpolation

````python
shrink_image = cv2.resize(img, (100, 100),
               interpolation = cv2.INTER_AREA)
plt.imshow(shrink_image)
print('Image size: ',shrink_image.shape)
````

````python
shrink_image = cv2.resize(img, (100, 100),
               interpolation = cv2.INTER_CUBIC)
plt.imshow(shrink_image)
print('Image size: ',shrink_image.shape)
````

````python
shrink_image = cv2.resize(img, (100, 100),
               interpolation = cv2.INTER_LINEAR)
plt.imshow(shrink_image)
print('Image size: ',shrink_image.shape)
````

````python
shrink_image = cv2.resize(img, (100, 100),
               interpolation = cv2.INTER_NEAREST)
plt.imshow(shrink_image)
print('Image size: ',shrink_image.shape)
````

### Add text to the image - putText()

````python
## choose font
font = cv2.FONT_HERSHEY_SCRIPT_COMPLEX#cv2.FONT_HERSHEY_COMPLEX
# FONT_HERSHEY_SIMPLEX
# FONT_HERSHEY_PLAIN
# FONT_HERSHEY_DUPLEX
# FONT_HERSHEY_COMPLEX
# FONT_HERSHEY_TRIPLEX
# FONT_HERSHEY_COMPLEX_SMALL 
# FONT_HERSHEY_SCRIPT_SIMPLEX
# FONT_HERSHEY_SCRIPT_COMPLEX
# FONT_ITALIC
blank_img = np.zeros(shape=(512,512,3))
cv2.putText(blank_img, text='sample text', org=(0,200), fontFace=font,
    fontScale=9, color=(0,1,1), thickness=1, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
````
