# Seaborn Figure Styles
The set_style() function allows you to set the theme of the plot. As per the latest updated version, below are the five themes available.

* Darkgrid
* Whitegrid
* Dark
* White
* Ticks


````python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
def sinplot(flip=1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5):
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sns.set_style("ticks")
sinplot()
plt.show()
````

## Removing Axes Spines
* In the white and ticks themes, we can remove the top and right axis spines using the despine() function.
* In the regular plots, we use left and bottom axes only. Using the despine() function, we can avoid the unnecessary right and top axes spines, which is not supported in Matplotlib.

````python
import numpy as np
from matplotlib import pyplot as plt
def sinplot(flip=1):
   x = np.linspace(0, 14, 100)
   for i in range(1, 5):
      plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
import seaborn as sb
sb.set_style("whitegrid")
sinplot()
sb.despine()
plt.show()
````

# Strip Plot
* It is similar to the scatter plot with one categorical variable
* It is used to understand the underlying distribution of the data 
* One axis represents the categorical variable and another represents the value corresponding to the categories

````python
import pandas as pd
df_titanic=pd.read_csv('data/Titanic_data.csv')
df_titanic.head(2)
````

## Check the distribution of age based on gender
* observation : The plot shows that, the maximum age of males is higher than of females


````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.stripplot(x='Sex', y='Age', data=df_titanic)
plt.title('Strip plot for age and sex')
plt.show()
````

## Add one more categorical variable to strip plot using the parameter, â€˜hueâ€™
* Observation : Proportion of female survivors is higher than males

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.stripplot(x='Sex', y='Age', hue='Survived',  data=df_titanic)
plt.title('survival based on age and sex')
plt.show()
````

# Distribution Plot
* It displays the distribution of the data
* It is a variation of histogram that uses kernel smoothing to plot values, allowing for smoother distributions by smoothing out the noise
* The distplot() method plots the histogram with a Kernel Density Estimator (KDE), which is a used to estimate the probability distribution function of a random variable

## Insurance claim distribution
* Observation : The plot shows the positive skewness of the â€˜claimâ€™ variable

````python
import pandas as pd
df=pd.read_csv('data/insurance_data.csv')
df.head(2)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(a=df['claim'])
plt.title('Claim distribution')
plt.show()
````

## Plot the distribution of Claim without the kernel density estimator (KDE)

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.distplot(a=df['claim'], color='olive', kde=False)
plt.title('Claim distribution')
plt.show()
````

# Joint Plot

# Violin Plot
* Violin Plots are a combination of the box plot with the kernel density estimates. So, these plots are easier to analyze and understand the distribution of the data.
* It is similar to a boxplot, that displays the kernel density estimator of the underlying distribution
* It shows the distribution of the quantitative data across categorical variables such that those distributions can be compared

* NOTE : The box plot is plotted inside the violin plot

````python
import pandas as pd
df_titanic=pd.read_csv('data/Titanic_data.csv')
df_titanic.head(2)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.violinplot(x='Sex', y='Age',  data=df_titanic)
plt.title('survival based on age and sex')
plt.show()
````

### Violin plot can be divided into two halves, where one half represents surviving while other half represents the non-surviving passenger
* Pass â€˜Trueâ€™ as value for the split parameter

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.violinplot(x='Sex', y='Age',  data=df_titanic, hue='Survived', split=True)
plt.title('survival based on age and sex')
plt.show()
````

# Heatmap
* A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented by the different colors
* Heatmap for correlation shows the correlation between the variables on each axis


````python
import pandas as pd
df_titanic=pd.read_csv('data/iris.csv')
df_titanic.head(2)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
#cbar- color bar
# annot If True, write the data value in each cell.
# cmap The mapping from data values to color space.
# linewidth line between cells
correlation=df_titanic.corr()
print(correlation)
sns.heatmap(data=correlation, annot=True, cmap='Reds', cbar=False, linewidth=0.5)
plt.title('Correlation Plot')
plt.show()
````

# Multiple plots with seaborn

````python
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset('penguins')
print(df.head())
fig, axs = plt.subplots(1,3, figsize=(20,5))
sns.countplot(x='species', data=df, ax=axs[0])
# sns.barplot(x='island', data=df, ax=axs[1])
sns.barplot(data=df, x="island", y="body_mass_g", ax=axs[1])
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", ax=axs[2])
axs[0].set_title('Chart 1')
axs[1].set_title('Chart 2')
axs[2].set_title('Chart 3')
plt.show()
````
