# Histogram
* It is used to represent the distribution of the numeric variable
* It is an estimate of the probability distribution of a continuous data
* One axis represents the variable in the form of bars and another represents the frequency each bar
* There are no gaps between the bars of the histogram

## Grouped Bar Plot
* Compare the marks of the students in R and Python

````python
import matplotlib.pyplot as plt
import numpy as np
python=[23,45,76,89,10]
R=[45,76,12,60,45]
rollno=np.arange(1,6)
names=['Ajay', 'Vinod', 'Kisor', 'Kumar', 'Madhav']
plt.bar(rollno, python,width=.2, label='Python')
plt.bar(rollno+0.2, R,width=.2, label='R')

plt.title('Marks distribution', color='b', fontsize=23)
plt.ylabel(' Makrs obtained', color='g', fontsize=20)
plt.xlabel('Student names', color='r', fontsize=20)

plt.xticks(rollno, names)
plt.legend()

plt.show()
````

## Stacked Bar Plot
* Compare the marks of the students in R and Python
* Plot the ‘R_marks’ above the ‘Python_marks’

````python
import matplotlib.pyplot as plt
import numpy as np
python=[23,45,76,89,10]
R=[45,76,12,60,45]
rollno=np.arange(1,6)
names=['Ajay', 'Vinod', 'Kisor', 'Kumar', 'Madhav']
plt.bar(rollno, python,width=.2, label='Python')
plt.bar(rollno, R,width=.2, bottom=python,label='R')

plt.title('Marks distribution', color='b', fontsize=23)
plt.ylabel(' Makrs obtained', color='g', fontsize=20)
plt.xlabel('Student names', color='r', fontsize=20)

plt.xticks(rollno, names)
plt.legend()

plt.show()
````

## Plot a pie plot to study the region wise patient distribution

````python
import pandas as pd
df=pd.read_csv('data/insurance_data_with_day.csv')
df.head(2)
````

````python
data=df['region'].value_counts().reset_index()
data.columns=['value', 'freq']
data
````

````python
import matplotlib.pyplot as plt

plt.pie(x=data['freq'], labels=data['value'], autopct='%.3f%%')

plt.title('Marks distribution', color='b', fontsize=23)

plt.show()
````

## Start Angle
* By default the plotting of the first wedge starts from the x-axis and moves counterclockwise

````python
import matplotlib.pyplot as plt

plt.pie(x=data['freq'], labels=data['value'], autopct='%.3f%%', startangle=90)

plt.title('Marks distribution', color='b', fontsize=23)

plt.show()
````

## Exploded Pie Plot
* The explode parameter allows you to do the wedges to stand out.
* If there are N pieces the N explod values must be specify
* The values for each explode must be between 0-1

````python
import matplotlib.pyplot as plt

plt.pie(x=data['freq'], labels=data['value'], autopct='%.3f%%', explode=(0.2,0,0,0))

plt.title('Marks distribution', color='b', fontsize=23)

plt.show()
````

# Matplotlib Subplots
* The matplotlib.pyplot.subplots method provides a way to create one or more subplots on a single figure.
* When number of rows and columns are passed, it returns a tuple (fig, ax).
* (fig, ax) represents a single figure, fig and and array of axes, ax.
* The fig is used as a container to hold all subplots
* The ax is a single object in case of one plot, array of axes in case of multiple plots

## Subplots
* The subplots(n rows,m columns) generates nXm space of subplots
* default is generates 1X1 subplot space

````python
import matplotlib.pyplot as plt

fig, ax=plt.subplots()
````

## 2X1 subplots

````python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,10)
sqrt=np.sqrt(x)
tan_x=np.tan(x)

fig, ax=plt.subplots(2,1)

ax[0].plot(x, sqrt, c='g')
ax[0].set_xlabel('X data')
ax[0].set_ylabel('sqrt data')
ax[0].set_title('X vs sqrt data')

ax[1].plot(x, tan_x, c='r')
ax[1].set_xlabel('X data')
ax[1].set_ylabel('tanx data')
ax[1].set_title('X vs tanx data')

fig.tight_layout(pad=2.0)

plt.show()
````

### Line plot of sinx and cosx in 3D

````python
import matplotlib.pyplot as pyplot
import numpy as np
from mpl_toolkits import mplot3d

fig = pyplot.figure(figsize = (10,10))
ax = pyplot.axes(projection = '3d')

#Mentioning all the three different axes.
z = np.linspace(0, 2, 800)
x = z * np.sin(80 * z)
y = z * np.cos(80 * z)

ax.plot3D(x, y, z, 'blue')
ax.set_title('3D plotting of line chart')
ax.set_xlabel('sin_x')
ax.set_ylabel('cos_x')
ax.set_zlabel('X data')
# Print the chart
pyplot.show()
````
