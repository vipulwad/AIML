* Rank the DataFrame by values in the column Sales using the parameter, method = ‘min’

# DataFrame Merge
* The merge() method concatenates the DataFrames based on one or more keys.
* If the column for join is not specified, the merge() method uses the overlapping column names as the keys.
* The merge types can be specified using the parameter, ‘how’. 
* By default the type is ‘inner’. Screenshot from 2023-03-04 12-03-01.png

```python
df1=pd.read_excel('data/Ecommerce_data.xlsx', sheet_name=0)
print(df1)
df2=pd.read_excel('data/Ecommerce_data.xlsx', sheet_name=1)
print(df2)
```

```python
df_inner = pd.merge(df1, df2, on='Cust_ID')
```

## Outer Merge

```python
df_outer = pd.merge(df1, df2, on='Cust_ID', how='outer')
```

```python
df3 =pd.concat([df_outer,df_inner], axis=0)
df3.drop_duplicates(keep=False)
```

## Right Merge

```python
pd.merge(df1, df2, on='Cust_ID', how='right')
```

## Left Merge

```python
pd.merge(df1, df2, on='Cust_ID', how='left')
```

```python

import pandas as pd
df = pd.DataFrame({'Courses': ["Spark","PySpark","Python","pandas","Java"],
                    'Fee' : [20000,25000,30000,24000,40000],
                    'Duration':['30day','40days','60days','55days','50days']})

df1 = pd.DataFrame({'subject': ["Java","PySpark","Python","pandas","Hyperion"],
                    'payment': [20000,25000,30000,24000,40000],
                    'Percentage':['10%','20%','25%','20%','10%']})
 
```

```python
df
```

```python
df1
```

```python

# Use pandas.merge() on multiple columns
df2 = pd.merge(df, df1, left_on=['Courses', 'Fee'], right_on=['subject', 'payment'])
print(df2)
```

# DataFrame Joins
* The join() method  combines the DataFrame rows from one or more tables based on key or index column
* Index of the first DataFrame should match to one of the column in the second DataFrame

* Types of Joins : The join types can be specified using the parameter, ‘how’
* If the type is not specified, by default it is ‘left’
Screenshot from 2023-03-03 00-50-03.png

```python
df1=pd.read_excel('data/customer.xlsx', sheet_name=0)
df1
```

```python
df2=pd.read_excel('data/customer.xlsx', sheet_name=1)
df2
```

## Join DataFrames using indexes.

```python
# lsuffix : Suffix to use from left frame’s overlapping columns.
# rsuffix : Suffix to use from right frame’s overlapping columns.

df1.join(df2, lsuffix='_customer', rsuffix='_order', how='inner')
```

## Join the DataFrames using column name

NOTE : If we want to join using the key columns, we need to set key to be the index in both df and other. The joined DataFrame will have key as its index.

```python
df1.set_index('Cust_ID').join(df2.set_index('Cust_ID'), on='Cust_ID', how='inner')
```
