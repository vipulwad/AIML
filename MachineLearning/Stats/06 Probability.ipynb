# Probability simulation

Q: Toss a coin and simulate the probability of N Trials

```python
import random
def coin_toss(num_toss):
    heads_count = 0
    tails_count = 0
    for i in range(num_toss):
        r=random.random()
        if r>0.5:
            #print('Heads')
            heads_count +=1
        else:
            #print('Tails')
            tails_count +=1
    return heads_count, tails_count
            


# Example usage
start = 1
stop = 5
exponent = 1

numbers = [10 ** (exponent * i) for i in range(start, stop + 1)]

print(numbers)
HEADS=[]
TAILS=[]

for number in numbers:
    print('Experiment : ', number)
    H,T=coin_toss(number)
    print("Heads: {}".format(H))
    print("Tails: {}".format(T))

    print('Probability of Head is {:.4f}'.format(H/number))
    print('Probability of Tail is {:.4f}'.format(T/number))
    HEADS.append(H/number)
    TAILS.append(T/number)
```

```python
import matplotlib.pyplot as plt

plt.plot(numbers, HEADS)
plt.plot(numbers, TAILS)
plt.show()
```

# Conditional Probability
Q. One coin and one dice is thrown. Find the probability of getting 4 from the dice given the coin shows head.

```python
import random

def conditional_probability_simulation(num_trials, condition, event):
    favorable_count = 0

    for _ in range(num_trials):
        # Simulate events based on the condition
        if condition():
            if event():
                favorable_count += 1

    # Calculate the conditional probability
    simulation_result = favorable_count / num_trials
    return simulation_result

# Example usage
num_trials = 10

def condition():
    # Define your condition here
    # Example: Coin toss is heads
    return random.choice([True, False])

def event():
    # Define your event here
    # Example: Rolling a 6-sided die and getting a 4
    return random.randint(1, 6) == 4

simulation_result = conditional_probability_simulation(num_trials, condition, event)
print("Simulation Result: {:.4f}".format(simulation_result))
```

# Probability mass function

The Probability Mass Function (PMF) is a function that describes the probability distribution of a discrete random variable. It assigns probabilities to each possible outcome or value of the random variable.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10,000 random integers between 1 and 100
random_data = np.random.randint(1, 10, size=10000)
random_data
```

```python
# Compute the PMF
unique_values, value_counts = np.unique(random_data, return_counts=True)
total_count = len(random_data)
pmf = value_counts / total_count
```

## Poisson Distribution
The Poisson distribution is a discrete probability distribution that is often used to model the occurrence of events in a fixed interval of time or space.
Example:

Number of Customer Arrivals per Hour--

2, 5, 3, 4, 1, 2, 3, 2, 4, 5, 3, 2, 4, 1, 3, 2, 3, 1, 2, 4


The Poisson distribution is a discrete probability distribution that models the number of events that occur within a fixed interval of time or space. It is often used to describe rare events or occurrences. The distribution is named after the French mathematician SimÃ©on Denis Poisson, who introduced it in the early 19th century.

The Poisson distribution is characterized by a single parameter, Î» (lambda), which represents the average rate of events occurring in the interval. This parameter determines both the mean (Î¼) and variance (Ïƒ^2) of the distribution, which are both equal to Î».

The probability mass function (PMF) of the Poisson distribution is given by:

PMF(k; Î») = (e^(-Î») * Î»^k) / k!

Where:

* k is the number of events (an integer)
* Î» is the average rate of events occurring in the interval

Let's consider an example to illustrate the Poisson distribution. Suppose you are a manager of a call center and you receive an average of 10 customer calls per hour. You can model the number of calls you receive within a specific hour using the Poisson distribution.

To simulate the number of calls, you can generate random numbers following a Poisson distribution with Î» = 10, representing the average rate of calls per hour. Let's generate and display a sample of 10 simulated call counts:

```python
import numpy as np

lambda_value = 10  # Average rate of calls per hour
sample_size = 10

# Generate random numbers following a Poisson distribution
calls = np.random.poisson(lambda_value, size=sample_size)

print("Simulated Number of Calls:")
print(calls)
```

## Binomial distribution

The Binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials. Each trial has two possible outcomes: success (usually represented as "1") with probability p, or failure (usually represented as "0") with probability q = 1 - p.

The probability mass function (PMF) of the Binomial distribution is given by:

PMF(k; n, p) = C(n, k) * p^k * q^(n-k)

Where:

* k is the number of successes (an integer)
* n is the number of trials (a positive integer)
* p is the probability of success in a single trial (0 â‰¤ p â‰¤ 1)
* q = 1 - p is the probability of failure in a single trial
* C(n, k) is the binomial coefficient, calculated as C(n, k) = n! / (k! * (n-k)!)

The mean (Î¼) and variance (Ïƒ^2) of the Binomial distribution are given by:
Î¼ = n * p
Ïƒ^2 = n * p * q

The Binomial distribution is commonly used to model events with two possible outcomes, such as coin flips, the success or failure of experiments, or the number of defective items in a sample.

# Bayes Theorem

Let's consider a simple example of classifying fruits based on their features: color (red or yellow) and shape (round or elongated). We have a dataset of 100 fruits with known labels: 60 apples and 40 bananas.

To classify a new fruit, we need to calculate the posterior probabilities for each class (apple and banana) given its features.

Assuming we have the following prior probabilities:

  P(apple) = 0.6 (60 apples out of 100 fruits)

  P(banana) = 0.4 (40 bananas out of 100 fruits)

We also need to calculate the likelihood probabilities based on the features.

For color:

  P(red|apple) = 0.8 (48 out of 60 apples are red)

  P(yellow|apple) = 0.2 (12 out of 60 apples are yellow)

  P(red|banana) = 0.4 (16 out of 40 bananas are red)

  P(yellow|banana) = 0.6 (24 out of 40 bananas are yellow)

For shape:

  P(round|apple) = 0.7 (42 out of 60 apples are round)

  P(elongated|apple) = 0.3 (18 out of 60 apples are elongated)

  P(round|banana) = 0.2 (8 out of 40 bananas are round)

  P(elongated|banana) = 0.8 (32 out of 40 bananas are elongated)

To classify a new fruit with features (red, elongated), we calculate the posterior probabilities for each class:

  P(apple|red, elongated) = (P(red|apple) * P(elongated|apple) * P(apple)) / P(red, elongated)

  P(banana|red, elongated) = (P(red|banana) * P(elongated|banana) * P(banana)) / P(red, elongated)

The denominator P(red, elongated) is a normalization factor calculated as:

  P(red, elongated) = (P(red|apple) * P(elongated|apple) * P(apple)) + (P(red|banana) * P(elongated|banana) * P(banana))

Comparing the values of P(apple|red, elongated) and P(banana|red, elongated), we can classify the fruit as either an apple or a banana based on the higher posterior probability.

```python
class NaiveBayesClassifier:
    def __init__(self):
        self.class_probabilities = {}
        self.feature_probabilities = {}

    def fit(self, X, y):
        classes = set(y)
        num_samples = len(X)

        # Calculate class probabilities
        for class_label in classes:
            class_count = sum(1 for label in y if label == class_label)
            self.class_probabilities[class_label] = class_count / num_samples

        # Calculate feature probabilities
        num_features = len(X[0])
        for feature_idx in range(num_features):
            self.feature_probabilities[feature_idx] = {}

            for class_label in classes:
                feature_values = set(X[:, feature_idx])
                self.feature_probabilities[feature_idx][class_label] = {}

                for value in feature_values:
                    matching_samples = X[(X[:, feature_idx] == value) & (y == class_label)]
                    num_matching_samples = len(matching_samples)

                    self.feature_probabilities[feature_idx][class_label][value] = num_matching_samples / class_count

    def predict(self, X):
        predictions = []

        for sample in X:
            posteriors = {}

            for class_label, class_prob in self.class_probabilities.items():
                likelihood = 1.0

                for feature_idx, feature_value in enumerate(sample):
                    likelihood *= self.feature_probabilities[feature_idx][class_label][feature_value]

                posterior = class_prob * likelihood
                posteriors[class_label] = posterior

            predicted_class = max(posteriors, key=posteriors.get)
            predictions.append(predicted_class)

        return predictions
```

## Approximating Area by Random Sampling

```python
import numpy as np
import matplotlib.pyplot as plt

def estimate_area(func, x_min, x_max, y_min, y_max, num_samples):
    count = 0

    for _ in range(num_samples):
        x = np.random.uniform(x_min, x_max)
        y = np.random.uniform(y_min, y_max)

        if func(x, y):
            count += 1

    area = (x_max - x_min) * (y_max - y_min) * (count / num_samples)
    return area

# Function to define the circle shape
def circle_func(x, y):
    return x**2 + y**2 <= 1

# Define the boundaries and number of samples
x_min, x_max = -1, 1
y_min, y_max = -1, 1
num_samples = 100

# Estimate the area
estimated_area = estimate_area(circle_func, x_min, x_max, y_min, y_max, num_samples)

# Plot the points inside and outside the circle
x_inside, y_inside = [], []
x_outside, y_outside = [], []

for _ in range(num_samples):
    x = np.random.uniform(x_min, x_max)
    y = np.random.uniform(y_min, y_max)

    if circle_func(x, y):
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)

# Plot the circle and sampled points
fig, ax = plt.subplots()
circle = plt.Circle((0, 0), 1, facecolor='None', edgecolor='blue')
ax.add_patch(circle)
ax.scatter(x_inside, y_inside, color='green', label='Inside Circle')
ax.scatter(x_outside, y_outside, color='red', label='Outside Circle')

# Set the aspect ratio and limits
ax.set_aspect('equal')
ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)

# Add a legend and title
ax.legend()
ax.set_title('Approximating Area of a Circle')

# Show the plot
plt.show()

print("Estimated Area:", estimated_area)

```

```python
import numpy as np

def draw_random_number(pmf_values, pmf_probs):
    # Compute cumulative probabilities
    cum_probs = np.cumsum(pmf_probs)

    # Generate a random number between 0 and 1
    random_num = np.random.random()

    # Find the index where the random number falls within the cumulative probabilities
    index = np.searchsorted(cum_probs, random_num)

    # Return the corresponding value from the PMF
    return pmf_values[index]

# Example usage
pmf_values = [1, 2, 3, 4]
pmf_probs = [0.2, 0.3, 0.4, 0.1]

random_number = draw_random_number(pmf_values, pmf_probs)
print("Random Number:", random_number)
```

```python
import random

def probability_simulation(num_trials, probability):
    success_count = 0

    for _ in range(num_trials):
        # Generate a random number between 0 and 1
        random_number = random.random()

        # Check if the random number is less than the given probability
        if random_number < probability:
            success_count += 1

    # Calculate the probability based on the number of successful trials
    simulation_result = success_count / num_trials
    return simulation_result

# Example usage
num_trials = 10000
probability = 0.7

simulation_result = probability_simulation(num_trials, probability)
print("Simulation Result: {:.4f}".format(simulation_result))
```
