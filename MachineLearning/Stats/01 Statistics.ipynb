### Using Numpy
```
numpy.histogram(data, bins, range)
```
* **data**	array or sequence of array to be plotted
* **bins**	int or sequence of str defines number of equal width bins in a range, default is 10
* **range**	optional parameter sets lower and upper range of bins

````python
x=np.random.randint(1,300, 100)
x
````

````python
hist, bins = np.histogram(x, bins=100, range=(0,100))
````

````python
print(hist, bins)
````

### cv2.calcHist() : in-built function in OpenCV to find the histogram.
```
cv2.calcHist(images, channels, mask, histSize, ranges)
```
* images : it is the source image of type uint8 or float32 represented as “[img]”.
* channels : it is the index of channel for which we calculate histogram. For grayscale image, its value is [0] and
color image, you can pass [0], [1] or [2] to calculate histogram of blue, green or red channel respectively.
* mask : mask image. To find histogram of full image, it is given as “None”.
* histSize : this represents our BIN count. For full scale, we pass [256].
* ranges : this is our RANGE. Normally, it is [0,256].

````python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread(img1)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
````

````python
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(histogram, color='k')
plt.show()
````

### Histogram for all 3 colors

````python
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(histogram, color='b')
plt.show()
````

````python
histogram = cv2.calcHist([image], [1], None, [256], [0, 256])
plt.plot(histogram, color='r')
plt.show()
````

````python
histogram = cv2.calcHist([image], [2], None, [256], [0, 256])
plt.plot(histogram, color='g')
plt.show()
````

### Histogram in one plot

````python

for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
    
plt.show()
````

### Masked histogram

````python
mask=image[450:600, 500:650] 
plt.imshow(mask)
#mask = np.zeros(mask[:2], dtype="uint8")
````

````python
for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([mask], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])
    
plt.show()
````

````python
mask = np.zeros(image.shape[:2], dtype="uint8")
mask[100:300, 200:400] = 255
````

## Measures of Central Tendency

* AM
* GM
* HM
* WM
* Median
* Weighted median
* Mode

````python
x=np.random.randint(1,10,5)
x
````

````python
# Arithmetic Mean
x.sum()/len(x)
````

````python
# Geometric Mean
x.prod()**(1/len(x))
````

````python
# Harmonic Mean
inv_x=1/x
print(inv_x)

len(x)/inv_x.sum()
````

````python
iris_df.head()
````

````python
iris_X=iris_df.iloc[:,[0,1,2,3]]
Y=np.cov(iris_X.T)

print(Y)
````
