# Text Preprocessing

## Process HTML page and extract the text

````python
import re
from bs4 import BeautifulSoup

# Sample HTML content
html_content = '''
<html>
<head><title>Sample Page</title></head>
<body>
<h1>Welcome to the Sample Page</h1>
<p>This is a sample paragraph with <a href="https://example.com">a link</a>.</p>
<div class="ad">Ad content here</div>
<p>Contact us at contact@example.com</p>
<p>Visit our website: https://example.com</p>
</body>
</html>
'''

# Step 1: Extract text from HTML
soup = BeautifulSoup(html_content, 'html.parser')
text = soup.get_text()
print(text)

````

````python
# Step 2: Remove HTML tags (Already done by BeautifulSoup's get_text method)

# Step 3: Remove advertisements and other non-text elements
# Assuming ads are within a specific HTML tag like <div class="ad">
text = re.sub(r'\bAd content here\b', '', text)

# Step 4: Normalize whitespace
text = re.sub(r'\s+', ' ', text).strip()

# Step 5: Remove non-text content such as URLs, emails
text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
text = re.sub(r'\S+@\S+', '', text)  # Remove email addresses

print(text)
````

## Download a wiki page and extract the contents


````python
import requests
from bs4 import BeautifulSoup
import re

# Step 1: Download the HTML page
url = 'https://en.wikipedia.org/wiki/Narendra_Modi'
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html_content = response.text
else:
    raise Exception(f"Failed to retrieve the webpage. Status code: {response.status_code}")
````

````python

# Step 2: Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Step 3: Extract the main content
# Wikipedia content is within the <div> with id "bodyContent"
content_div = soup.find('div', id='bodyContent')

# Get text from the content div
text = content_div.get_text()
print(text)
````

````python

# Step 4: Clean the text
# Remove [edit] tags and reference numbers like [1], [2], etc.
text = re.sub(r'\[edit\]', '', text)
text = re.sub(r'\[\d+\]', '', text)

# Normalize whitespace
text = re.sub(r'\s+', ' ', text).strip()

# Output the cleaned text
print(text)
````

# Subword tokenization
Subword tokenization is an important technique in Natural Language Processing (NLP) used to handle the issue of out-of-vocabulary (OOV) words and to provide more flexible representations of words. It is particularly useful in languages with rich morphology or when dealing with large vocabularies.

Popular subword tokenization techniques include Byte Pair Encoding (BPE), WordPiece, and SentencePiece. Here’s a brief overview of each:
**Byte Pair Encoding (BPE)**

**BPE** is a simple yet powerful tokenization technique that iteratively merges the most frequent pair of characters or subwords. This approach allows for a variable-length tokenization, capturing common subword units.


**WordPiece** is similar to BPE but is primarily used in models like BERT. It iteratively builds a vocabulary of subword units by splitting words into smaller pieces based on their frequency.
SentencePiece

**SentencePiece** is a more general framework for subword tokenization that treats the input text as a sequence of Unicode characters. It can use algorithms like BPE or Unigram Language Model to generate subword units.

## Byte Pair Encoding (BPE)
BPE is a popular subword tokenization technique used in NLP to handle large vocabularies and out-of-vocabulary words. Here's an example of how to implement BPE in Python using the tokenizers library by Hugging Face, which provides a simple and efficient way to work with BPE.

```
pip install tokenizers
```

````python
from tokenizers import Tokenizer, models, trainers, pre_tokenizers, decoders, processors

# Sample text data
sample_texts = [
    "This is an example of text used to train the BPE tokenizer.",
    "Byte Pair Encoding is useful for subword tokenization.",
    "It helps in handling large vocabularies efficiently."
]

# Save sample text to a file for training
with open("sample_texts.txt", "w") as f:
    for text in sample_texts:
        f.write(text + "\n")

# Initialize a BPE model
tokenizer = Tokenizer(models.BPE())

# Initialize a trainer for the BPE model
trainer = trainers.BpeTrainer(vocab_size=1000, special_tokens=["<unk>", "<s>", "</s>", "<pad>", "<mask>"])

# Initialize a pre-tokenizer (whitespace splitting)
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()

# Train the tokenizer on the sample text file
tokenizer.train(files=["sample_texts.txt"], trainer=trainer)

# Test the tokenizer
encoded = tokenizer.encode("This is an example of BPE tokenization.")
print("Encoded tokens:", encoded.tokens)

# Decode the tokens back to the original sentence
decoded = tokenizer.decode(encoded.ids)
print("Decoded sentence:", decoded)

# Save the tokenizer model
tokenizer.save("bpe_tokenizer.json")

# Load the tokenizer model
loaded_tokenizer = Tokenizer.from_file("bpe_tokenizer.json")

# Test the loaded tokenizer
encoded_loaded = loaded_tokenizer.encode("Byte Pair Encoding example.")
print("Encoded tokens with loaded tokenizer:", encoded_loaded.tokens)
````

## WordPiece
WordPiece tokenization is a subword tokenization technique used in models like BERT. It splits words into subword units, ensuring that the most frequent subwords are kept together, which helps in handling out-of-vocabulary words effectively.

````python
from tokenizers import Tokenizer, models, trainers, pre_tokenizers, processors

# Sample text data
sample_texts = [
    "This is an example of text used to train the WordPiece tokenizer.",
    "WordPiece is useful for subword tokenization.",
    "It helps in handling large vocabularies efficiently."
]

# Save sample text to a file for training
with open("sample_texts.txt", "w") as f:
    for text in sample_texts:
        f.write(text + "\n")

# Initialize a WordPiece model
tokenizer = Tokenizer(models.WordPiece(unk_token="[UNK]"))

# Initialize a trainer for the WordPiece model
trainer = trainers.WordPieceTrainer(vocab_size=1000, special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])

# Initialize a pre-tokenizer (whitespace splitting)
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()

# Train the tokenizer on the sample text file
tokenizer.train(files=["sample_texts.txt"], trainer=trainer)

# Test the tokenizer
encoded = tokenizer.encode("This is an example of WordPiece tokenization.")
print("Encoded tokens:", encoded.tokens)

# Decode the tokens back to the original sentence
decoded = tokenizer.decode(encoded.ids)
print("Decoded sentence:", decoded)

# Save the tokenizer model
tokenizer.save("wordpiece_tokenizer.json")

# Load the tokenizer model
loaded_tokenizer = Tokenizer.from_file("wordpiece_tokenizer.json")

# Test the loaded tokenizer
encoded_loaded = loaded_tokenizer.encode("WordPiece tokenization example.")
print("Encoded tokens with loaded tokenizer:", encoded_loaded.tokens)
````

## SentencePiece
SentencePiece is a powerful and flexible tokenizer library that can handle various tokenization tasks, including subword tokenization. It's developed by Google and is widely used for tasks like training BPE or Unigram language models. Here's an example of how to use SentencePiece for subword tokenization in Python.
```
pip install sentencepiece

```

````python
import sentencepiece as spm

# Sample text data
sample_texts = [
    "This is an example of text used to train the SentencePiece model.",
    "SentencePiece is useful for subword tokenization.",
    "It helps in handling large vocabularies efficiently."
]

# Save sample text to a file for training
with open("sample_texts.txt", "w") as f:
    for text in sample_texts:
        f.write(text + "\n")

# Adjust the vocabulary size to match the dataset
vocab_size = 50  # Set to a value <= 51

# Train the SentencePiece model
spm.SentencePieceTrainer.Train(f'--input=sample_texts.txt --model_prefix=m --vocab_size={vocab_size}')

# Load the trained SentencePiece model
sp = spm.SentencePieceProcessor()
sp.load('m.model')

# Test the tokenizer
sample_sentence = "This is an example of SentencePiece tokenization."
encoded_pieces = sp.encode_as_pieces(sample_sentence)
encoded_ids = sp.encode_as_ids(sample_sentence)

# Decode back to the original sentence
decoded_sentence = sp.decode_pieces(encoded_pieces)

print("Original Sentence:", sample_sentence)
print("Encoded Pieces:", encoded_pieces)
print("Encoded IDs:", encoded_ids)
print("Decoded Sentence:", decoded_sentence)
````

## Stemming
* Stemming is a text processing task in which you reduce words to their root, which is the core part of a word. For example, the words “helping” and “helper” share the root “help.” Stemming allows you to zero in on the basic meaning of a word rather than all the details of how it’s being used.
* There are several different algorithms for stemming, including the Porter stemmer, Snowball stemmer, and the Lancaster stemmer.
  * The Porter stemmer is the most widely used algorithm, and it is based on a set of heuristics that are used to remove common suffixes from words.
  * The Snowball stemmer is a more advanced algorithm that is based on the Porter stemmer, but it also supports several other languages in addition to English.
  * The Lancaster stemmer is a more aggressive stemmer and it is less accurate than the Porter stemmer and Snowball stemmer.

* Stemming can be useful for several natural language processing tasks such as text classification, information retrieval, and text summarization

````python
# single word stemming
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
ps = PorterStemmer()
example_words = ["python","pythoner","pythoning","pythoned","pythonly"]
for w in example_words:
    print(ps.stem(w))
````

````python
from nltk.stem import SnowballStemmer
````

## Lemmatization
* In contrast to stemming, lemmatization is a lot more powerful. It looks beyond word reduction and considers a language’s full vocabulary to apply a morphological analysis to words, aiming to remove inflectional endings only and to return the base or dictionary form of a word, which is known as the lemma
* lemmatizing reduces words to their core meaning, but it will give you a complete English word that makes sense on its own instead of just a fragment of a word like 'discoveri'.

````python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("discoveries")
# Discovery discovered many discoveries
````

````python
string_for_lemmatizing = "The friends of DeSoto love scarves."
words = word_tokenize(string_for_lemmatizing)
words
````

````python
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
lemmatized_words
````
