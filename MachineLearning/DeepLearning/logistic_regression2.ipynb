```python
import numpy as np
```

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
features, labels = make_blobs(n_samples=300, centers=2,
                       cluster_std=0.60, random_state=0)
plt.scatter(features[:, 0], features[:, 1], c = labels,cmap='rainbow');
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,  y_test=train_test_split(features, labels, test_size=0.2)
X_train.shape, y_train.shape
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
#model=LinearRegression()
model=LogisticRegression()
model.fit(X_train, y_train)
pred_y=model.predict(X_test)
from sklearn.metrics import accuracy_score, f1_score

print("Accuracy", accuracy_score(y_test, pred_y))
print("f1_score", f1_score(y_test, pred_y))
trained_w=model.coef_
trained_b=model.intercept_
```

```python
plt.scatter(features[:, 0], features[:, 1], c=labels)
# Plot the decision boundary (line where predictions are 0.5)
boundary_x = np.array([min(features[:, 0]), max(features[:, 0])])
boundary_y = (-trained_b - trained_w[0][0] * boundary_x) / trained_w[0][1]
plt.plot(boundary_x, boundary_y, label='Decision Boundary', color='red')

plt.title('Logistic Regression on 2D Dataset')
plt.legend()
plt.show()
```
