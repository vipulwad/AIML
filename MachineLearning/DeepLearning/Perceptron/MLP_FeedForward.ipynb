## Q.
As shown in the image a simple neural network contains D=2 input node Q=3 hidden node and M=2 output nodes.
1. Take a random input $X_1=\{x_1, x_2\}$
2. Take $w_1, w_2$ random matrix as weights and $b_1,  b_2$ random vector as bias.
3. Perform Feedforward for the given network
  
  a. Pass the $X_1$ data from D to Q
   
   $U=w_1 X_1+b_1$

   $H=σ(U)$

  b. Pass the $H$ data from Q to M
   
   $V=w_2 H+b_2$
   
   $Y=σ(V)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
X1 = np.random.rand(2)
W1 = np.random.rand(3,2)
B1 = np.random.rand(3)
```

```python
print("Input data X1:", X1)
print("Wright Matrix W1:", W1)
print("Bias Vector B1:", B1)


print("X1 shape:", X1.shape)
print("W1 shape:", W1.shape)
print("B1 shape:", B1.shape)
```

```python
U = np.matmul(W1,X1.reshape(2,1))
U = np.add(U,B1.reshape(3,1))
H = sigmoid(U)
print(U, H)
```

```python
W2 = np.random.rand(2,3)
B2 = np.random.rand(2,1)
```

```python
print("Input data H:", H)
print("Wright Matrix W2:", W2)
print("Bias Vector B2:", B2)

print("H shape:", H.shape)
print("W2 shape:", W2.shape)
print("B2 shape:", B2.shape)
```

```python
V = np.matmul(W2,H)
V = np.add(V,B2)
O = sigmoid(V)
print(V, O)
```
