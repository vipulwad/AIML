# Unsupervised Learning

* Parametric Clustering Algorithms:

  **Assumption**: Parametric algorithms make specific assumptions about the form of the underlying data distribution.

  **Example**: Gaussian Mixture Models (GMM), K-Means.

  **Characteristic**: Assumes clusters follow a particular parametric distribution (e.g., Gaussian) and aims to estimate the parameters of that distribution.

* Generic Clustering Algorithms:

  **Assumption**: Nonparametric algorithms make minimal assumptions about the underlying data distribution and often do not require specifying the number of clusters in advance.

  **Example**: Mean Shift.

  **Characteristic**: Particularly useful when the number of clusters is unknown or when dealing with complex, irregularly shaped clusters. They adapt to the inherent structure of the data.

## Key Differences

* Distribution Assumptions:

  * Parametric: Assumes a specific parametric form for the underlying data distribution.
  * Generic: Makes minimal assumptions about the data distribution and often does not require specifying the number of clusters.
* Flexibility:

  * Parametric: Less flexible due to strict distribution assumptions.
  * Generic:  Highly flexible, adapting to the data structure without strong assumptions.

* Number of Clusters:

  * Parametric: Often requires specifying the number of clusters in advance.
  * Generic: Often does not require specifying the number of clusters, allowing the algorithm to discover the structure autonomously.

* Shape of Clusters:

 * Parametric: Assumes a specific shape for clusters (e.g., spherical in K-Means, Gaussian in GMM).
 * Generic: Adapts to various shapes and sizes of clusters without assuming a particular form.

## Summary
In summary, parametric clustering algorithms make strong assumptions about the underlying data distribution, generic clustering make minimal assumptions and often do not require specifying the number of clusters in advance. The choice of algorithm depends on the characteristics of the data and the goals of the analysis.

## What is estimation theory?
Estimation theory is a branch of statistics that deals with the process of estimating the characteristics of a population based on a sample of data drawn from that population. The goal is to make informed predictions or inferences about the population parameters using statistical methods.

or Estimation theory involves methods for estimating unknown parameters or characteristics of a population based on a sample of data. This is crucial in statistics, where we often deal with incomplete information and want to infer characteristics of a larger group based on a smaller group that we observe.

* Parametric Estimation:

  * Assumption: Assumes that the data follows a specific distribution or model. It specifies the form of the distribution and the parameters that characterize it.
  
  * Example: Assuming the data follows a normal distribution, you estimate the mean and standard deviation as parameters.

  * Flexibility: More rigid, as it requires a strong assumption about the data distribution. The accuracy of the estimation depends on the correctness of the assumed distribution.

  * Common Methods: Maximum Likelihood Estimation (MLE).

* Nonparametric Estimation:

  * Assumption: Makes fewer or no assumptions about the specific form of the underlying distribution. It focuses on estimating functions rather than parameters.

  * Example: Using the median or kernel density estimation without assuming a specific distribution.

  * Flexibility: More flexible, especially in situations where the true data distribution is unknown or difficult to specify. Can handle a wide range of data distributions.

  * Common Methods: Kernel Density Estimation, Median, Rank-based methods.



# Parametric estimation
Parametric estimation involves making specific assumptions about the underlying distribution of the data and then using observed data to estimate the parameters of that assumed distribution.
1. Choose a Model : Select a specific probability distribution or statistical model that you believe represents the data well. Common distributions include normal (Gaussian), exponential, Poisson, etc.
2. Define Parameters : Identify the parameters of the chosen distribution. Parameters are the characteristics that determine the shape and location of the distribution. For example, in a normal distribution, parameters include the mean and standard deviation.
3. Collect Data : Gather a sample of data from the population of interest. The sample should be representative of the larger population.
4. Estimate Parameters : Use the collected data to estimate the parameters of the chosen distribution. The estimation process aims to find the values for the parameters that make the assumed distribution most likely to have produced the observed data.

# Nonparametric estimation

Nonparametric estimation involves estimating characteristics of a population without making specific assumptions about the underlying probability distribution. nonparametric estimation is typically done:

1. Collect Data:

  Gather a sample of data from the population of interest. The sample should be representative of the larger population.

2. Select a Nonparametric Method:

  Choose a nonparametric method based on the specific characteristics you want to estimate. Common nonparametric methods include kernel density estimation, median estimation, rank-based methods, etc.

## Kernel Density Estimation (KDE)
1. Kernel Selection:

  Choose a kernel function. The kernel determines the shape of the smoothing function around each data point.
2. Bandwidth Selection:

  Select a bandwidth parameter. The bandwidth controls the width of the kernel and influences the smoothness of the estimated density.

3. Smooth the Data:

  Place a kernel around each data point and sum them up to create a smooth estimate of the underlying probability density function.
  The formula for the kernel density estimate at a point x might look like this

  $f(x)= \frac{1}{nh} \sum_{i=1}^{n} K \left( \frac{x-x_i}{h} \right)$

4. Sum Kernel Contributions:

  Sum up the contributions from all data points to create the overall KDE.

5. Mode finding :

  Calculate the numerical derivative of the KDE curve and identifying points where the derivative changes from positive to negative. This technique is used to locate local maxima, which correspond to peaks in the KDE curve and, in the context of probability density estimation, represent modes in the distribution.

### KDE in 1D data

```python
import numpy as np
import matplotlib.pyplot as plt
# Create a synthetic dataset (you can replace this with your own data)
np.random.seed(42)
```

```python
np.random.normal(10, 3, (10,2,3))
```

```python

data = np.concatenate([np.random.normal(0, 1, 30), np.random.normal(14, 10, 30), np.random.normal(24, 26, 30)])
print(data.shape)
# Reshape the data to fit the scikit-learn API
data = data.reshape(-1, 1)
print(data.shape)
```

```python
# Plot the original data
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', label='Histogram')
plt.scatter(data, np.zeros_like(data), marker='x', color='red')
plt.xlabel('Data Points')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
from scipy.stats import norm
# Generate points for plotting
x_plot = np.linspace(min(data), max(data), 1000).reshape(-1, 1)
for ix,point in enumerate(data):
    y = norm.pdf(x_plot, loc=point, scale=0.1)  # Adjust the scale as needed
    plt.scatter(point, 0)
    plt.plot(x_plot, y, linestyle='dashed', linewidth=0.5, color='black')
    # if ix==3:
    #   break
```

```python
from sklearn.neighbors import KernelDensity
# Create a KDE model
# kernel{gaussian, tophat, epanechnikov, exponential, linear, cosine}, default=gaussian
# bandwidth= float or {scott, silverman}, default=1.0
# The bandwidth of the kernel. If bandwidth is a float, it defines the bandwidth of the kernel. If bandwidth is a string, one of the estimation methods is implemented.
kde = KernelDensity(bandwidth=0.75, kernel='gaussian')
kde.fit(data)
```

```python
# Generate points for plotting
x_plot = np.linspace(min(data), max(data), 1000).reshape(-1, 1)

# Evaluate the KDE model on the generated points
log_dens = kde.score_samples(x_plot)
# Plot the KDE estimate
plt.plot(x_plot, np.exp(log_dens), color='red', label='Kernel Density Estimate')


plt.title('KDE for 1D Data with Modes Identified')
plt.xlabel('Data Points')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Find peaks (modes) using a simple derivative approach
# 1. Calculate the Numerical Derivative
# Use the np.diff function to calculate the differences between consecutive elements
# in the array representing the KDE curve. This effectively computes a
# numerical approximation of the derivative.

derivative = np.diff(np.exp(log_dens))
print(derivative)
```

```python
# 2. Locate points where the derivative changes from positive to negative.
# These points indicate the presence of a peak or local maximum in the KDE curve.

peaks = (derivative[:-1] > 0) & (derivative[1:] < 0)
print(peaks)
```

```python
# 3. Extract the x-values corresponding to the identified peaks.
#  These x-values represent the locations of modes in the KDE.

peak_values = x_plot[1:-1][peaks]
print(peak_values)
```

```python
# Plot modes
plt.scatter(peak_values, np.exp(kde.score_samples(peak_values.reshape(-1, 1))), color='green', marker='o', label='Modes')
plt.plot(x_plot, np.exp(log_dens), color='red', label='Kernel Density Estimate')
plt.title('KDE for 1D Data with Modes Identified')
plt.xlabel('Data Points')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python

plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', label='Histogram')
# Plot modes
plt.scatter(peak_values, np.exp(kde.score_samples(peak_values.reshape(-1, 1))), color='yellow', marker='o', label='Modes')
plt.plot(x_plot, np.exp(log_dens), color='red', label='Kernel Density Estimate')
plt.title('KDE for 1D Data with Modes Identified')
plt.xlabel('Data Points')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## KDE for 2d Data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
from mpl_toolkits.mplot3d import Axes3D

# Function to generate 2D data with two peaks
def generate_bimodal_data():
    np.random.seed(42)
    data1 = np.random.multivariate_normal(mean=[-3, -3], cov=[[0.3, 0.5], [0.5, 1]], size=50)
    data2 = np.random.multivariate_normal(mean=[-2, 2], cov=[[1, -0.5], [-0.5, 1]], size=300)
    return np.concatenate([data1, data2])

# Generate bimodal 2D data
data_2d_bimodal = generate_bimodal_data()
print(data_2d_bimodal.shape)

```

```python
# Create a KDE model for 2D data
kde_2d_bimodal = KernelDensity(bandwidth=0.5, kernel='gaussian')
kde_2d_bimodal.fit(data_2d_bimodal)

# Generate points for plotting in 2D
x_plot = np.linspace(min(data_2d_bimodal[:, 0]), max(data_2d_bimodal[:, 0]), 100)
y_plot = np.linspace(min(data_2d_bimodal[:, 1]), max(data_2d_bimodal[:, 1]), 100)
x_plot, y_plot = np.meshgrid(x_plot, y_plot)
xy_plot = np.column_stack([x_plot.ravel(), y_plot.ravel()])

# Evaluate the KDE model on the generated points
log_dens_2d_bimodal = kde_2d_bimodal.score_samples(xy_plot)

# Reshape the results for contour plotting
log_dens_2d_bimodal = log_dens_2d_bimodal.reshape(x_plot.shape)

# Plot the original 2D data points
plt.scatter(data_2d_bimodal[:, 0], data_2d_bimodal[:, 1], alpha=0.5, color='blue', label='Data Points')

# Plot contours of the KDE estimate
plt.contour(x_plot, y_plot, np.exp(log_dens_2d_bimodal), levels=10, colors='red', linewidths=2)

plt.title('KDE for Bimodal 2D Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# Visualize KDE peaks in 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Create meshgrid for 3D surface plot
x_plot_3d, y_plot_3d = np.meshgrid(x_plot[0, :], y_plot[:, 0])  # Use 1D arrays for meshgrid
z_plot_3d = np.exp(log_dens_2d_bimodal)

# Plot 3D surface
ax.plot_surface(x_plot_3d, y_plot_3d, z_plot_3d, cmap='viridis', edgecolor='k')

ax.set_title('3D Visualization of KDE Peaks')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Density')

plt.show()
```

## KDE and seeking mode using Gradent Ascent for 2D dataset

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic 2D data with three clusters
np.random.seed(42)
data1 = np.random.multivariate_normal([2, 2], [[1, 0.5], [0.5, 1]], 100)
data2 = np.random.multivariate_normal([6, 6], [[1, 0], [0, 1]], 100)
data3 = np.random.multivariate_normal([10, 2], [[1, -0.5], [-0.5, 1]], 100)
data = np.vstack([data1, data2, data3])
data.shape
```

```python
plt.scatter(data[:,0], data[:,1])
```

```python
from scipy.stats import gaussian_kde
from scipy.optimize import minimize
```

```python
# Create a KDE object for the 2D data
kde = gaussian_kde(data.T)
```

```python
# Define the negative sum of KDE function (to perform minimization)
neg_kde_sum = lambda x: -kde(x).sum()
```

```python
# Choose starting points for optimization
start_points = np.array([[2, 2], [6, 6], [10, 2], [20,-20], [30,1]])
```

```python
# Perform optimization (gradient ascent) and store intermediate results
modes = []
optimization_steps = []
for start_point in start_points:
    result = minimize(neg_kde_sum, start_point, bounds=[(min(data[:, 0]), max(data[:, 0])), (min(data[:, 1]), max(data[:, 1]))])
    modes.append(result.x)
    optimization_steps.append(result.x)
    # print(result)

# Convert modes to numpy array for easier indexing
modes = np.array(modes)
modes
```

```python
# Plot the original data, KDE contours, and highlight mode locations
plt.scatter(data[:, 0], data[:, 1], alpha=0.5, label='Data')
# plt.contourf(*np.meshgrid(np.linspace(min(data[:, 0]), max(data[:, 0]), 100), np.linspace(min(data[:, 1]), max(data[:, 1]), 100)), kde(np.meshgrid(np.linspace(min(data[:, 0]), max(data[:, 0]), 100), np.linspace(min(data[:, 1]), max(data[:, 1]), 100))), cmap='viridis', levels=10, alpha=0.5, label='KDE')
plt.scatter(modes[:, 0], modes[:, 1], color='red', marker='o', s=100, label='Modes')
plt.scatter(np.array(optimization_steps)[:, 0], np.array(optimization_steps)[:, 1], marker='x', color='green', label='Optimization Steps')
plt.legend()
plt.title("Multidimensional KDE with Mode Locations")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
