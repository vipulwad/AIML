# Parametric Clustering
Parametric clustering is a type of clustering algorithm used in machine learning and data analysis. Clustering is the process of grouping similar data points together based on certain features or characteristics. Parametric clustering methods assume that the data can be described by a specific statistical distribution or model, and they seek to find the parameters of that distribution that best fit the data.

Popular parametric clustering algorithms include

* K-Means: Although it's not strictly a parametric method, K-means is often used with Gaussian distributions and can be considered a simple parametric clustering approach.

* Gaussian Mixture Models (GMM): GMM assumes that data points are generated by a mixture of Gaussian distributions and estimates the parameters of these distributions.

# Adaptive Thresholding

```python
import matplotlib.pyplot as plt
import numpy as np
dataset = [-1,27,31,2,59,3,61,34,0,12]
T0 = 3
def adap_thres(dataset, ti, epsilon=0.01):
    if len(dataset)==0:
        return
    tf = 100000
    while abs(tf-ti)>epsilon:
        sl = []
        sh = []
        for i in dataset:
            if i > ti:
                sh.append(i)
            else:
                sl.append(i)
        if len(sl)==0:
            ul = 0
        else:
            ul = sum(sl)/len(sl)
        if len(sh)==0:
            uh = 0
        else:
            uh= sum(sh)/len(sh)
        tf = ti
        ti = (ul + uh)/2
        print(sh)
        print(sl)
        print(tf)
        print("*"*10)
        x=sl
        y=np.zeros(len(sl))
        plt.scatter(sl,y, label='less then threshhold', marker='^')

        x=sh
        y=np.zeros(len(sh))
        plt.scatter(x,y, label='greater then threshhold')

        plt.scatter(ti,0, label='threshhold', marker='x')
        plt.legend()
        plt.show()
    return tf
```

```python
adap_thres(dataset, 3)
```

# Result after various thresholding schemes

```python
img = cv2.medianBlur(org_img,5) ## Remove noise using median Blur

## 2nd argument - threshold, 3rd Argument - Value assigned if pixel is greater than threshold
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

## threshold value is the mean of neighbourhood area
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,9,2)

## threshold value is the weighted sum of neighbourhood values where weights are a gaussian window.
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,9,2)

titles = ['After MedianFiltering','Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']

images = [img, th1, th2, th3]

plt.figure(figsize=(12,10))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.tight_layout()
plt.show()
```
