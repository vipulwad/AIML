# Logistic regression
Logistic regression is a statistical method used for binary classification, which means it is employed to predict the probability of an instance belonging to one of two classes. Despite its name, logistic regression is a classification algorithm rather than a regression algorithm. It is widely used in various fields, including medicine, finance, and social sciences.



# How logistic regression comes in picture of discrimenent function?
Discriminant functions and logistic regression are closely related in the context of classification problems. Discriminant functions aim to discriminate between different classes, and logistic regression is one of the methods to achieve this discrimination, especially in binary classification scenarios.

Let's explore how logistic regression fits into the framework of discriminant functions:

1. **Linear Discriminant Function:**
   - In the context of binary classification, a linear discriminant function is essentially a linear combination of the input features with associated weights.
   - Mathematically, it can be represented as: $ \text{discriminant function} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n $
   - Here, $ \beta_0, \beta_1, \ldots, \beta_n $ are the parameters (weights), and $ x_1, x_2, \ldots, x_n $ are the input features.

2. **Logistic Regression as a Discriminant Function:**
   - Logistic regression uses a linear combination of input features, similar to the linear discriminant function, but it applies the logistic (sigmoid) function to the result to produce a probability.
   - The logistic function ensures that the output is bounded between 0 and 1, which is suitable for representing probabilities.
   - Mathematically, the logistic regression equation is: $ P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n)}} $
   - Here, $ P(Y=1) $ is the probability of belonging to class 1.

3. **Decision Boundary:**
   - The decision boundary in logistic regression is determined by the threshold of 0.5. If the predicted probability is greater than or equal to 0.5, the instance is classified as belonging to class 1; otherwise, it is classified as belonging to class 0.

4. **Maximum Likelihood Estimation:**
   - Logistic regression is derived from the principles of maximum likelihood estimation (MLE), where the parameters are estimated to maximize the likelihood of observing the given set of outcomes given the input features.

In summary, logistic regression can be considered a form of discriminant function in the context of binary classification. It uses a linear combination of input features and applies a logistic function to make predictions and assign probabilities to instances belonging to different classes. The decision boundary is determined based on a threshold, and the model parameters are estimated through maximum likelihood estimation.

Here are the key components and concepts associated with logistic regression:

1. **Binary Outcome:**
   - Logistic regression is used when the dependent variable is binary, meaning it has only two possible outcomes (usually coded as 0 and 1).
   - The goal is to model the probability that an instance belongs to one of the two classes.

2. **Sigmoid Function (Logistic Function):**
   - The logistic regression model uses the logistic function (sigmoid function) to map the linear combination of input features and their associated weights into a probability score between 0 and 1.
   - The sigmoid function is defined as: $ \sigma(z) = \frac{1}{1 + e^{-z}} $
   - Here, \(z\) is the linear combination of the input features and weights.

3. **Linear Combination of Features:**
   - The linear combination of input features and their corresponding weights is represented as $z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n$, where $\beta_0, \beta_1, \ldots, \beta_n$ are the model parameters (weights), and $x_1, x_2, \ldots, x_n$ are the input features.

4. **Probability and Decision Boundary:**
   - The output of logistic regression is a probability score between 0 and 1. If the predicted probability is greater than or equal to a chosen threshold (commonly 0.5), the instance is assigned to one class; otherwise, it is assigned to the other class.
   - The decision boundary is the value of \(z\) for which the probability is exactly 0.5.

5. **Maximum Likelihood Estimation:**
   - Logistic regression estimates its parameters (weights) using maximum likelihood estimation (MLE). The objective is to maximize the likelihood of observing the given set of outcomes given the input features.

6. **Cost Function:**
   - The model is trained by minimizing a cost function, which is often the negative log-likelihood loss. This involves adjusting the weights to improve the model's predictions.



## Data generation

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Number of data points for each class
num_samples = 100

# Generate random features for Class 0
class_0_features = np.random.normal(loc=0, scale=1, size=(num_samples, 2))

# Generate random features for Class 1
class_1_features = np.random.normal(loc=3, scale=1, size=(num_samples, 2))

# Create labels for the classes (0 for Class 0, 1 for Class 1)
class_0_labels = np.zeros((num_samples, 1))
class_1_labels = np.ones((num_samples, 1))

# Concatenate features and labels for each class
class_0_data = np.concatenate((class_0_features, class_0_labels), axis=1)
class_1_data = np.concatenate((class_1_features, class_1_labels), axis=1)

# Concatenate data from both classes to form the final dataset
full_dataset = np.concatenate((class_0_data, class_1_data), axis=0)

# Shuffle the dataset
np.random.shuffle(full_dataset)

# Separate features and labels
features = full_dataset[:, :-1]
labels = full_dataset[:, -1]


# Plot the dataset
plt.scatter(class_0_features[:, 0], class_0_features[:, 1], label='Class 0')
plt.scatter(class_1_features[:, 0], class_1_features[:, 1], label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Random 2D Dataset')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    """Sigmoid (logistic) activation function."""
    return 1 / (1 + np.exp(-z))
```

```python
def initialize_parameters(num_features):
    """Initialize weights and bias."""
    w = np.zeros((num_features, 1))
    b = 0
    return w, b
```

```python
# w, b=initialize_parameters(20)
# w.shape
```

```python
def compute_cost(X, y, w, b):
    """Compute the logistic regression cost (negative log-likelihood)."""
    m = len(y)
    h = sigmoid(np.dot(X, w) + b)
    cost = -(1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost
```

```python
def gradient_descent(X, y, w, b, learning_rate, num_iterations):
    """Gradient descent optimization to update weights and bias."""
    m = len(y)

    for _ in range(num_iterations):
        # print(X.shape, w.shape)
        h = sigmoid(np.dot(X, w) + b)
        # dw = (1/m) * np.dot(X.T, (h - y))
        dw = (1/m) * np.sum(np.dot(X.T, (h - y)))
        db = (1/m) * np.sum(h - y)

        w -= learning_rate * dw
        b -= learning_rate * db

    return w, b
```

```python
def train_logistic_regression(X, y, learning_rate=0.01, num_iterations=1000):
    """Train the logistic regression model."""
    w, b = initialize_parameters(X.shape[1])

    # Add a column of ones to X for the bias term
    # X = np.column_stack((np.ones((X.shape[0], 1)), X))

    w, b = gradient_descent(X, y, w, b, learning_rate, num_iterations)

    return w, b
```

```python
def predict(X, w, b):
    """Make predictions using the logistic regression model."""
    # X = np.column_stack((np.ones((X.shape[0], 1)), X))
    predictions = sigmoid(np.dot(X, w) + b)
    return (predictions >= 0.5).astype(int)
```

```python
# Use the previously generated random 2D dataset
X_train = features
y_train = labels.reshape(-1, 1)

# Train logistic regression model
trained_w, trained_b = train_logistic_regression(X_train, y_train,
                                                 learning_rate=0.01,
                                                 num_iterations=10000)

# Make predictions on the training set
predictions_train = predict(X_train, trained_w, trained_b)
print("Weights and Bias", trained_w, trained_b)
# Calculate accuracy on the training set
accuracy_train = np.mean(predictions_train == y_train)
print("Accuracy on the training set:", accuracy_train)
```

```python
# Plot the decision boundary
plt.scatter(class_0_features[:, 0], class_0_features[:, 1], label='Class 0')
plt.scatter(class_1_features[:, 0], class_1_features[:, 1], label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the decision boundary (line where predictions are 0.5)
boundary_x = np.array([min(features[:, 0]), max(features[:, 0])])
boundary_y = (-trained_b - trained_w[0] * boundary_x) / trained_w[1]
plt.plot(boundary_x, boundary_y, label='Decision Boundary', color='red')

plt.title('Logistic Regression on 2D Dataset')
plt.legend()
plt.show()
```

## Iris Dataset with Logistic Regression

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
X = iris.data[iris.target<2]
y = iris.target[iris.target<2]
```

```python
X_train,X_test, y_train,  y_test=train_test_split(X[:, 0:2], y, test_size=0.2)
X_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train, y_train)
pred_y=model.predict(X_test)
from sklearn.metrics import accuracy_score, f1_score

print("Accuracy", accuracy_score(y_test, pred_y))
print("f1_score", f1_score(y_test, pred_y))
```

```python
trained_w=model.coef_
trained_b=model.intercept_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
# Plot the decision boundary (line where predictions are 0.5)
boundary_x = np.array([min(X[:, 0]), max(X[:, 0])])
boundary_y = (-trained_b - trained_w[0][0] * boundary_x) / trained_w[0][1]
plt.plot(boundary_x, boundary_y, label='Decision Boundary', color='red')
```
