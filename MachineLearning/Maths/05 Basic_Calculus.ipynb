```python
## import the libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Write Mathematical functions in python

$y$ = 1 / (1 + $x^2$)       $x$ = [-5, 5]

```python
def myfunc_1(x):
    return 1/(1+x**2)
```

```python
x = np.linspace(-5,5,100)
y = myfunc_1(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

$x^2$ + $y^2$ = 1 $x$ = [-1,1]

```python
def myfunc_2(x):
    return (1-x**2)**0.5
```

```python
x = np.linspace(-1,1,100)
y = myfunc_2(x)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y)
plt.plot(x,-y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

y = sin(x)  x=[- $\pi$,$\pi$]

```python
np.pi
```

```python
def myfunc_3(x):
    return np.sin(x)
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = myfunc_3(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

y = 1/(1+e^(-x))  x = [0,10]

```python
def myfunc_4(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(0,10,100)
y = myfunc_4(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y,'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Continuous and Discontinuous function

y = sin(x)  at x =$\pi$/2

```python
def myfunc_5(x):
    return np.sin(x)
h = 0.01
```

```python
y_lhl = myfunc_5(np.pi/2 - h)   ## Left hand limit
y_lhl
```

```python
y_rhl = myfunc_5(np.pi/2 + h)  ## Right hand limit
y_rhl
```

```python
y = myfunc_5(np.pi/2)  ## function at c
y
```
