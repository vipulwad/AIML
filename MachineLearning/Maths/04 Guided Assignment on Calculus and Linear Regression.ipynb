# Differentiation


The derivative of a function of a real variable measures the
sensitivity to change of the function value (output value) with respect
to a change in its argument (input value).

## Numerical Differentiation

### Differentiation using Python

**Activation Functions:**
* Linear Function:  y= ax
* Sigmoid Function: A = 1/(1 + e-x)
* Tanh: tanh(x) = 2/(1 + e-2x) - 1
* RELU: A(x) = max(0,x)
* Parametrised RELU: A(x) = max(px,x) {p usually less than 0ne.}
* Leaky RELU: p = 0.1
* Swish: f(x) = x/(1-e^-x)

```python
# Plot activation functions
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5, 100)
sigmoid = 1/(1+np.exp(-x))

def p_relu(x, p= 0.1):
    return np.maximum(p*x, x)

relu = p_relu(x)
prelu = p_relu(x, 0.1)
plt.plot(x, sigmoid, label = "sigmoid")
#plt.plot(x, relu, label= "relu")
#plt.plot(x,prelu, label="prelu")
plt.legend()
```

```python
# Using NumPy and Python to find derivative
def der_func(func, value, h=0.01):
    return (func(value+h) - func(value-h))/(2*h)
derv_relu = [der_func(p_relu, i) for i in x]
plt.plot(x, derv_relu)
```

```python
# Using Sympy to find derivative
# symbols, Symbol, diff, f.diff, lamdify 
from sympy import *
x = Symbol('x')
y = x**3 + 10*x**2 + 100
print(y)
y_diff = y.diff(x)
print(y_diff)
func = lambdify(x, y)
print(func)
print(func(0.1))
der_func = lambdify(x, y_diff)
print(der_func(1))
```

```python
# Effect of "h" in numerical differentiation
x = np.linspace(-5, 5, 5)

def der_func(func, value, h=0.01):
    return (func(value+h) - func(value-h))/(2*h)

def sigm(x):
    return x**6 + 20*x**2 + 10

y0 = [der_func(sigm, i, 0.001) for i in x]
y1 = [der_func(sigm, i, 0.01) for i in x]
y2 = [der_func(sigm, i, 0.1) for i in x]
y3 = [der_func(sigm, i, 1) for i in x]

print(y0)
print(y1)
print(y2)
print(y3)
```

# Taylor Series & Maclaurin Theorem

### Few common examples:


```python
# Effect of each term
import math

def exp_ts(x, n=20):
    result = 0
    for i in range(n):
        result += (x**i)/math.factorial(i)
    return result

x = 20

print(np.exp(x))
print(exp_ts(x, 5))
print(exp_ts(x, 15))
print(exp_ts(x, 25))
print(exp_ts(x, 35))
print(exp_ts(x, 45))
print(exp_ts(x, 60))
print(exp_ts(x, 80))
print(exp_ts(x, 100))
```

# Partial Derivatives

A partial derivative of a function of several variables is its derivative
with respect to one of those variables, with the others held constant.

```python
# Calulation of partial derivatives using Sympy
x, y = symbols('x y')
func = exp(x)*(y**5)
print(func)
print(func.diff(y))
print(func.diff(x,y))
```

```python
# Calulation of partial derivatives using Sympy
x, y = symbols('x y')
func = x**10 + 2*(y**2)*(x**3) + 10*x
print(func)
print(func.diff(y))
print(func.diff(x,y))
```
