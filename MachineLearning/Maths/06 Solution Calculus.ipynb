# Numerical Differentiation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The function generates data
t=np.linspace(-6,10, 100)
f_x=np.sin(t)
dt=0.2
```

```python
# Exact derivatives
dfdt=np.cos(t)
```

```python
# Ploting function and derivatives
plt.plot(t, f_x, '--', c='r')
plt.plot(t, dfdt, c='g')
plt.legend(['Function', 'Exact derivatives'])
plt.xlim(-5,9)
plt.ylim(-1.5,1.5)
plt.show()
```

## Forward Approximation
$\hat f^{\prime}= \frac {f(t+Ît)-f(a)}{Ît}$

```python
# Forward Approximation
dfdtF=(np.sin(t+dt)-np.sin(t))/dt
```

## Backward Approximation
$\hat f^{\prime}= \frac {f(a) - f(t-Ît)}{Ît}$

```python
# Backward Approximation
dfdtB=(np.sin(t)-np.sin(t-dt))/dt
```

## Central Approximation
$\hat f^{\prime}= \frac {f(t+Ît) - f(t-Ît)}{2Ît}$

```python
# Central Approximation
dfdtC=(np.sin(t+dt)-np.sin(t-dt))/(2*dt)
```

```python
# Ploting derivatives
plt.plot(t, dfdt, c='g')
plt.plot(t, dfdtF, '--' , c='b')
plt.plot(t, dfdtB, '-.' , c='k')
plt.plot(t, dfdtC, ':',  c='r')
plt.legend(['Exact derivatives', 'Forward', 'Backward', 'Central'])
plt.xlim(-6,10)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
delta_t=np.linspace(0.0002, 2, 10)
delta_t
from sklearn.metrics import mean_squared_error
Err=[]
for dt in delta_t:
  dfdtF=(np.sin(t+dt)-np.sin(t))/dt
  # print(dfdtF)
  # print(dfdt)
  error=mean_squared_error(dfdt, dfdtF, squared=True)
  Err.append(error)
  # break
plt.scatter(delta_t, Err)
```

Display the surface plot and contour plot of the above function in the
search space given by $S_{min} = [x_{min}, y_{min}]^T = [â10, â10]^T$ and $S_{max} = [x_{max}, y_{max}]^T = [10, 10]^T$


```python
def calc(x, y, a, b,c,d,e):
  x_=x+b
  y_=y+c
  X=(x_**2)/d
  Y=(y_**2)/e
  X_add_Y=X+Y
  z=a*np.exp(-X_add_Y)
  return z
```

### Random point inside a 3D ellipse

```python
import numpy as np
def random_point_ellipsoid(a,b,c):
    u = np.random.rand()
    v = np.random.rand()
    theta = u * 2.0 * np.pi
    phi = np.arccos(2.0 * v - 1.0)
    sinTheta = np.sin(theta);
    cosTheta = np.cos(theta);
    sinPhi = np.sin(phi);
    cosPhi = np.cos(phi);
    rx = a * sinPhi * cosTheta;
    ry = b * sinPhi * sinTheta;
    rz = c * cosPhi;
    return rx, ry, rz
p=[]
for j in range(5):
  for i in range(100):
    x,y,z=random_point_ellipsoid(np.random.randint(0,9),
                                 np.random.randint(0,9),
                                 np.random.randint(0,9))

    x=np.random.randint(-400, 400, 1)+x
    y=np.random.randint(-400, 400, 1)+y
    z=np.random.randint(-400, 400, 1)+z
    p.append([x, y, z])

p=np.asarray(p)
```

```python
fig = plt.figure(figsize = (5,5))
ax = plt.axes(projection='3d')

ax.scatter(p[:,0], p[:,1], p[:,2], c = 'r', s = 5)
# ax.view_init(0,180)
plt.show()
```
